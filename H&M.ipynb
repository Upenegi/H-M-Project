```python
#Another way of finding missing data by creating function:

def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis = 1, keys = ['Total', 'Percent'])

missing_data(df)
```

there is no null or missing values

```python
##what and how many of the product types are there in different product group? 

# .nunique() method in Pandas is used to count the number of unique values in a DataFrame or Series. 
#It returns the count of distinct, non-null values present in the data.

temp = articles_df.groupby(["product_group_name"])["product_type_name"].nunique()
df = pd.DataFrame({'Product Group': temp.index,
                   'Product Types': temp.values
                  })
df = df.sort_values(['Product Types'], ascending = False)
plt.figure(figsize = (8,6))
plt.title('Number of Product Types per each Product Group')
sns.set_color_codes("muted")
s = sns.barplot(x = 'Product Group', y="Product Types", data=df)
s.set_xticklabels(s.get_xticklabels(),rotation=90)
locs, labels = plt.xticks()
plt.show()
```

```python
##How many articles are there in different product group

temp = articles_df.groupby(['product_group_name'])['article_id'].nunique()
df = pd.DataFrame({'Product Group': temp.index,
                   'Articles': temp.values
                  })
df = df.sort_values(['Articles'], ascending = False)
plt.figure(figsize = (8,6))
plt.title('Number of articles in different product groups')
sns.set_color_codes('pastel')
s = sns.barplot(x = 'Product Group', y = 'Articles', data = df)
s.set_xticklabels(s.get_xticklabels(), rotation = 90)
locs, labels = plt.xticks()
plt.show()
```

```python
def unique_values(data):
    total = data.count()
    tt = pd.DataFrame(total)
    tt.columns = ["Total"]
    uniques = []
    for col in data.columns:
        unique = data[col].nunique()
        uniques.append(unique)
    tt['Uniques'] = uniques
    return tt
```
